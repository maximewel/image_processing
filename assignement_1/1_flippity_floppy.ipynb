```python
#Imports

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plt_format(image: any) -> any:
    """Function that uses the cvtcolor function from openCV to map the openCV-decoded image (B,G,R) into a regular (R,G,B) format by re-ordering the pixel channels"""
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

# Original
Import image, show original

```python
image = cv2.imread('./in/Lenna.png') 
plt.imshow(plt_format(image))
```

# Eventual feature extraction
Metadata reading to help understan the image

```python
y,x,z = image.shape
print(f"Image is {x}*{y} pixels, pixels have {z} channels")
```

# Horizontal flip
openCV matrixes are not [x,y] - they are [line, col] so the horizontal flip applies on the second dimension

```python
#Reverse every horiz pixel using python slices (Start:end:step). Equivalent to a for loop on the dimension.
imgHorizFlip = image[:, ::-1]
plt.imshow(plt_format(imgHorizFlip))
```

# Vertical flip
Similarly, the [line, col] format of the matrixes means the Y axis is the first dimension of the matrix

```python
imgVerticalFlip = image[::-1, :]
plt.imshow(plt_format(imgVerticalFlip))
```

# diagonal flip
for this one I wanted to demonstrate that it can work with both the same slicing method or a simple double loop

```python
imgBothFlip = image[::-1, ::-1]
plt.imshow(plt_format(imgBothFlip))
```

```python
#Blank canvas
imgBothFlipLoop = np.zeros(image.shape, np.uint8)

#Fill it during double loop
for i in range(x):
    for j in range(y):
        imgBothFlipLoop[i, j] = image[x-1-i, y-1-j]

plt.imshow(plt_format(imgBothFlipLoop))
```

# Closout
End the assignement by writing the resulting images into this folder

```python
if input("Type anything to save images, ignore to skip"):
    cv2.imwrite("./out/horizontal.png", imgHorizFlip)
    cv2.imwrite("./out/vertical.png", imgVerticalFlip)
    cv2.imwrite("./out/both.png", imgBothFlip)
    cv2.imwrite("./out/both_loop.png", imgBothFlipLoop)
```
